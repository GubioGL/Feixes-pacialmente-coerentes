```python
from LightPipes import *
import matplotlib.pyplot as plt
```

```python
# Definindo as propriedade do campo.
## Tamanho de cada pixel
dx = 10*mm
## Tamanho da imagem da simulação
Tamanho = 2**8
## Comprimento de onda
Comprimento_de_onda = 620*nm
## Criando o campo
F = Begin(dx,Comprimento_de_onda,Tamanho)
# Cintura do feixe
Cintura  = 2*mm
## Moldando a fase e intensidade
F = GaussLaguerre(F,Cintura, p=1, l=1, ecs=0)  # Psi Eq.()
F = Forvard(F, 0.2*m)
I = Intensity(0,F)

plt.imshow(I,cmap="jet");
plt.axis('off')
plt.show()
```
